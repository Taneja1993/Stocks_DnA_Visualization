Importing the us stocks dataset (__us_stocks.csv__) and selecting the Adjusted Close (__Adj Close__) for all six stocks from __31-12-2015__ (incl.) until __31-12-2018__ (incl.). Saving the DataFrame in the variable __stocks__ and work through the next features with this DataFrame!

```python
import pandas as pd
import numpy as np
```

```python
stocks = pd.read_csv("us_stocks.csv")
stocks
```

```python
stocks = pd.read_csv("us_stocks.csv", index_col = [0], header = [0,1], parse_dates = [0])
stocks
```

```python
stocks = stocks.loc["31-12-2015":"31-12-2018" ,"Adj Close"]
```

```python
stocks
```

__Visualizing__ the Adjusted Close of the six stocks __over time__!

```python
import matplotlib.pyplot as plt
plt.style.use("seaborn")
stocks.plot(figsize = (12,8))
plt.show()
```

__Visualizing__ the Adjusted Close of the six stocks __over time__, starting each at the __Base Value 100__!

```python
stocks.iloc[0]
```

```python
stocks = stocks.div(stocks.iloc[0]).mul(100)
```

```python
stocks.plot(figsize = (12,8))
plt.show()
```

Calculating __monthly returns__ for all six stocks and creating a new __returns__ DataFrame!

```python
stocks_monthly = stocks.resample("BM").last()
```

```python
stocks_monthly.pct_change().dropna()
```

```python
returns = stocks_monthly.pct_change().dropna()
returns
```

Calculating the __annualized mean return__ and the __annualized standard deviation__ of returns for all six stocks!

```python
returns_ann = returns.describe().T.loc[:, ["mean", "std"]]
```

```python
returns_ann["mean"] = returns_ann["mean"] * 12
returns_ann["std"] = returns_ann["std"] * np.sqrt(12)
returns_ann
```

__Visualizing risk/return__ of all six stocks in a (total) risk / return framework. (annualized std on the x-axis, annualized mean returns on the y-axis). <br>
Identifying the __best performing__ stock and the __worst performing__ stock (in terms of __risk / return__)

```python
returns_ann.plot(kind = "scatter", x = "std", y = "mean", figsize = (12, 8), s = 50, fontsize = 15, xlim = (0.1, 0.3), ylim = (0, 0.25))
for i in returns_ann.index:
    plt.annotate(i, xy=(returns_ann.loc[i, "std"]+0.002, returns_ann.loc[i, "mean"]+0.002), size = 15)
plt.xlabel("ann. Risk(std)", fontsize = 15)
plt.ylabel("ann. Return", fontsize = 15)
plt.title("Risk/Return", fontsize = 20)
plt.show()
```
